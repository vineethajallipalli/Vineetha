```python
import pandas as pd

# Load the CSV file into a DataFrame
sales_data = pd.read_csv('/content/Amazon Sale Report.csv')

# Display the first few rows of the dataset to check the structure
sales_data.head()
```

```python
# Check for missing values
print(sales_data.isnull().sum())
```

```python
# View column names
print(sales_data.columns)

# Check data types of each column
print(sales_data.dtypes)
```

```python
print(sales_data.columns)
```

```python
sales_data.columns = sales_data.columns.str.strip()
```

```python
category_sales = sales_data.groupby('Category')['Amount'].sum().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the data by 'Category' and sum the 'Amount' (representing sales)
category_sales = sales_data.groupby('Category')['Amount'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Amount', y='Category', data=category_sales, palette='viridis')
plt.title('Sales by Product Category')
plt.show()
```

```python
fulfillment_sales = sales_data.groupby('Fulfilment')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Amount', y='Fulfilment', data=fulfillment_sales, palette='viridis')
plt.title('Sales by Fulfillment Type')
plt.show()
```

```python
sales_data.columns = sales_data.columns.str.strip()
```

```python
sales_channel_sales = sales_data.groupby('Sales Channel')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Amount', y='Sales Channel', data=sales_channel_sales, palette='viridis')
plt.title('Sales by Sales Channel')
plt.show()
```

```python
size_sales = sales_data.groupby('Size')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Amount', y='Size', data=size_sales, palette='viridis')
plt.title('Sales by Product Size')
plt.show()
```

```python
country_sales = sales_data.groupby('ship-country')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Amount', y='ship-country', data=country_sales, palette='viridis')
plt.title('Sales by Country')
plt.show()
```
